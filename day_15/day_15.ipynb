```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
# import seaborn as sns
%matplotlib inline
# mpl.style.use("seaborn") # make our plot look pretty

x = np.random.rand(100)
y = np.sin(x)
```

```python
plt.plot(x)
plt.title("Some line plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.xticks()
```

```python
plt.tight_layout() #  forece plot into specified
plt.savefig("analysis.svg")
```

```python
# using the OOP approach

fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1,0.1,0.8,1])
axes.plot(x)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
plt.savefig("line_plot.svg")
```
